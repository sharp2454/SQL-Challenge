```python
#import employees table
employees = pd.read_sql('select * from employees', connection)
employees.head()
```

```python
#import salaries table
salaries = pd.read_sql('select * from salaries', connection)
salaries.head()
```

```python
#import department manager table
dept_managers = pd.read_sql('select * from dept_managers', connection)
dept_managers.head()
```

```python
#import titles table
titles = pd.read_sql('select * from titles', connection)
titles
```

```python
#import department employees table
dept_employ = pd.read_sql('select * from dept_employ', connection)
dept_employ.head()
```

```python
#create a merged dataframe with salaries and employee ids
employ_salary = employees.merge(salaries, on='employ_no')
employ_salary.head()
```

```python
#create a merged dataframe with employee title ids and salaries
employ_titles = employ_salary.merge(titles, on='employ_title_id')
employ_titles.head()
```

```python
#create a dataframe with titles and salaries
employee_salaries_df = employ_titles[['title', 'salary']]
employee_salaries_df.head()
```

```python
#groupby titles
employee_salaries_df.groupby('title')['salary'].mean()
employee_salaries_df
```
